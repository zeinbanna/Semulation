```python
import matplotlib.pyplot as plt

simulationLength=20
population=1000
growthRate=0.5
timestep=0.005
numIterations=int(simulationLength/timestep)
k=growthRate*timestep
M=1000
t=0
pop=[]
for i in range (numIterations+1):
    population=population+k*(1-(population/M))*population
    pop.append(population)
    if(t==10): 
        M=2000
    t=i*timestep
  


plt.plot(pop)
plt.show()   
```
