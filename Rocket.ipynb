```python
import matplotlib.pyplot as plt
import math

initial_mass=5000
rocket_mass=1000
burnout_time=60

mass=initial_mass

change_in_mass=(initial_mass-rocket_mass)/burnout_time

specific_impulse=200

g= -9.81

velocity=0

Δt=0.1

position=0

simulationLength=200

numIterations=int(simulationLength/Δt)

mass=5000

position=0

t=0

velocityList=[]
positionList=[]

for i in range(numIterations+1):
    
    if (t>burnout_time):
         change_in_mass=0
    
    mass= mass-change_in_mass * Δt
    
    acceleration=-(specific_impulse*g *change_in_mass )/mass+g
    
    velocity+= acceleration * Δt # معدل السرعة
    
    position =position + velocity *  Δt # الطول النهائي
    
    velocityList.append(velocity)
    
    positionList.append(position/1000)
    
    t= Δt * i
   

plt.plot(positionList,color="blue") 
plt.ylabel("position")
plt.xlabel("Time")
plt.show()    
    


plt.plot(velocityList,color="red") 
plt.ylabel("velocity")
plt.xlabel("Time")
plt.show()    
```
