```python
import matplotlib.pyplot as plt
import math


acceleration_due_to_gravity= -9.81 # تسارع الجاذبية الارضية
mass=0.2 # الكتلة
weight=acceleration_due_to_gravity * mass  # الوزن

spring_constant=10 # ثابت النابض


init_displacement=0.3 # طوله نتيجة الشد 


Δt=0.02 # التغير في الزمن

simulationLength=3

numIterations=int(simulationLength/Δt)

velocity=0 # السرعة الابتدائية - النابض ساكن

unweighted_length= 1.4962 # طوله الاصلي

weight_displacement = -weight/spring_constant # طوله نتيجة الوزن 

init_displacement=0.3 # طوله بعد الشد 

length= unweighted_length + weight_displacement + init_displacement # الطول الكلي

LengthList=[]

for i in range(numIterations+1):
    
    restoring_spring_force = - spring_constant * (length - unweighted_length) # قوة النابض
    
    total_force =restoring_spring_force + weight # القوة الكلية 
    
    acceleration = total_force/mass # التسارع 
    
    velocity+= acceleration * Δt # معدل السرعة
    
    length =length + velocity *  Δt # الطول النهائي
    
    LengthList.append(length); 
    
    
plt.plot(LengthList,color="blue")  
plt.xlabel("Time")
plt.ylabel("Length")
plt.show()

```
