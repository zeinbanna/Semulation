```python
import matplotlib.pyplot as plt
import math



BTS_population = 15

BTS_birth_fraction = 1

BTS_death_proportionality_constant = 0.20

WTS_population= 20

WTS_birth_fraction = 1

WTS_death_proportionality_constant = 0.27


Δt=0.01

simulationLength=5

numIterations=int(simulationLength/Δt)

BTSList=[]
WTSList=[]
    
for i in range(numIterations+1):
    
    BTS_births = BTS_birth_fraction * BTS_population 
    BTS_deaths = (BTS_death_proportionality_constant * WTS_population) *BTS_population
    WTS_births = WTS_population * WTS_birth_fraction
    WTS_deaths = (WTS_death_proportionality_constant * BTS_population) *WTS_population

    BTS_population+=(BTS_births - BTS_deaths)*Δt
    WTS_population-=(WTS_births - WTS_deaths)*Δt
    
    BTSList.append(BTS_population)
    WTSList.append(WTS_population)
    
    
    
plt.plot(BTSList,color="red")
    
plt.plot(WTSList,color="green") 

plt.ylabel("Population")
plt.xlabel("Time")
plt.show()      
    
    
```
